```python
import numpy as np
import torch
from src.data.window import create_windows
from src.data.dataloader import create_dataloaders
from src.models.build import build_model
from src.training.train import train_model
from src.evaluation.metrics import smape


def train_best_model(best_row, train_data, val_data, test_data):

    device = torch.device("mps" if torch.backends.mps.is_available() else "cpu")

   
    # 1 Merge train + val
    train_val = np.vstack([train_data, val_data])

    seq_len = int(best_row["seq_len"])
    pred_len = int(best_row["pred_len"])

    
    # 2 Create windows
    X_train, Y_train = create_windows(train_val, seq_len, pred_len)
    X_test, Y_test = create_windows(test_data, seq_len, pred_len)

    
    # 3 Residual targets
    Y_train_res = Y_train - X_train[:, -1:, :]
    Y_test_res = Y_test - X_test[:, -1:, :]

    
    # 4 Dataloaders
    train_loader, _ = create_dataloaders(
        X_train,
        Y_train_res,
        X_train,
        Y_train_res,
        batch_size=16
    )


    # 5 Build config
    config = {
        "model_type": best_row["model_type"],
        "seq_len": seq_len,
        "pred_len": pred_len,
        "n_heads": int(best_row["n_heads"]),
        "dropout": float(best_row["dropout"]),
        "d_model": int(best_row["d_model"]),
        "enc_layers": int(best_row["enc_layers"]),
        "input_dim": train_val.shape[1],
        "epochs": 20,
        "save_path": f"results/{best_row['model_type']}_final.pth",
        "lr":0.0001,
        "patience":5,
        "residual":True,
        "univariate":False

    }

    
    # 6 Build model
    model = build_model(config, device)

    # 7 Train
    train_model(
        model,
        train_loader,
        train_loader,
        config,
        device
    )


    # 8 Test evaluation
    model.eval()
    test_smape = 0

    with torch.no_grad():

        for i in range(len(X_test)):

            x = torch.tensor(X_test[i:i+1], dtype=torch.float32).to(device)
            y_res = torch.tensor(Y_test_res[i:i+1], dtype=torch.float32).to(device)

            residual_pred = model(x)

            last_value = x[:, -1:, :]
            final_pred = last_value + residual_pred
            y_true = last_value + y_res

            test_smape += smape(y_true, final_pred).item()

    test_smape /= len(X_test)

    return test_smape
```

```python
from src.data.load_data import load_exchange_data

train_data,val_data,test_data=load_exchange_data(univariate=False)
```

```python
transformer_test=train_best_model(
    best_transformer,
    train_data,
    val_data,
    test_data
)

informer_test=train_best_model(
    best_informer,
    train_data,
    val_data,
    test_data
)
autoformer_test=train_best_model(
    best_autoformer,
    train_data,
    val_data.
    test_data
)
```
