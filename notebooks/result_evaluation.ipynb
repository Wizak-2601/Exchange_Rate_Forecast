```python
import numpy as np
import torch
from src.data.window import create_windows
from src.data.dataloader import create_dataloaders
from src.models.build import build_model
from src.training.train import train_model
from src.evaluation.metrics import smape
from src.data.normalization import fit_scaler,transform_data
import pandas as pd

def train_best_model(best_row, train_data, val_data, test_data):

    device = torch.device("mps" if torch.backends.mps.is_available() else "cpu")

   
    # 1 Merge train + val
    train_val = np.vstack([train_data, val_data])

    # --- normalization ---
    scaler = fit_scaler(train_val)
    train_val = transform_data(train_val, scaler)
    test_data = transform_data(test_data, scaler)

    seq_len = int(best_row["seq_len"])
    pred_len = int(best_row["pred_len"])

    
    # 2 Create windows
    X_train, Y_train = create_windows(train_val, seq_len, pred_len)
    X_test, Y_test = create_windows(test_data, seq_len, pred_len)

    
    # 3 Residual targets
    Y_train_res = Y_train - X_train[:, -1:, :]
    Y_test_res = Y_test - X_test[:, -1:, :]

    
    # 4 Dataloaders
    train_loader, _ = create_dataloaders(
        X_train,
        Y_train_res,
        X_train,
        Y_train_res,
        batch_size=16
    )


    # 5 Build config
    config = {
        "model_type": best_row["model_type"],
        "seq_len": seq_len,
        "pred_len": pred_len,
        "n_heads": int(best_row["n_heads"]),
        "dropout": float(best_row["dropout"]),
        "d_model": int(best_row["d_model"]),
        "enc_layers": int(best_row["enc_layers"]),
        "input_dim": train_val.shape[1],
        "epochs": 20,
        "save_path": f"results/{best_row['model_type']}_final.pth",
        "lr":0.0001,
        "patience":5,
        "residual":True,
        "univariate":False,
        "kernel_size":23,
        "use_scaling":True
    }

    
    # 6 Build model
    model = build_model(config, device)

    # 7 Train
    train_model(
        model,
        train_loader,
        train_loader,
        config,
        device
    )
    # saving the model
    model_name=config["model_type"]
    save_path=f"../results/final_models/{model_name}_final.pth"
    torch.save(
    model.state_dict(),
    save_path
    )
    

    # 8 Test evaluation
    model.eval()
    test_smape = 0
    actual_list=[]
    pred_list=[]
    with torch.no_grad():

        for i in range(len(X_test)):

            x = torch.tensor(X_test[i:i+1], dtype=torch.float32).to(device)
            y_res = torch.tensor(Y_test_res[i:i+1], dtype=torch.float32).to(device)

            residual_pred = model(x)

            last_value = x[:, -1:, :]
            final_pred = last_value + residual_pred
            y_true = last_value + y_res

            test_smape += smape(y_true, final_pred).item()
            actual_list.append(y_true.cpu().numpy().flatten())
            pred_list.append(final_pred.cpu().numpy().flatten())
    test_smape /= len(X_test)
    actual=np.concatenate(actual_list)
    pred=np.concatenate(pred_list)

    
    forecast_df = pd.DataFrame({
        "time": np.arange(len(actual)),
        "actual": actual,
        config["model_type"]: pred
    })

    # simple uncertainty band
    forecast_df[f"{config['model_type']}_upper"] = pred + 0.02
    forecast_df[f"{config['model_type']}_lower"] = pred - 0.02

    save_path = f"results/{config['model_type']}_forecast.csv"

    forecast_df.to_csv(save_path, index=False)

    print(f"Forecast saved to {save_path}")
    


    return test_smape
```

```python
from src.data.load_data import load_exchange_data

train_data,val_data,test_data=load_exchange_data(univariate=False)
```

```python
transformer_test=train_best_model(
    best_transformer,
    train_data,
    val_data,
    test_data
)

informer_test=train_best_model(
    best_informer,
    train_data,
    val_data,
    test_data
)
autoformer_test=train_best_model(
    best_autoformer,
    train_data,
    val_data,
    test_data
)
```

### Feature normalization was introduced using StandardScaler fit on the training data and applied to validation/test sets. While normalization stabilized training, it did not significantly reduce the validation-to-test generalization gap. This suggests that the gap is primarily caused by temporal distribution shift in the exchange-rate dataset rather than feature scale differences.
