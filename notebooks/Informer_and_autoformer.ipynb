### Here the models are tested for different values of Prediction Length, starting with informer and then followed by Autoformer

```python
#  informer seq-variable, pred-variable
for i in [48,96,192,256]:
    for j in [24,96,199,366]:
        config = load_config(
        "src/configs/base.yaml",
        overrides={
            "model_type": "informer",
            "seq_len": i,
            "pred_len": j,
            "d_model":60,
            "n_heads": 3,

        }
    )

        res = run_experiment(config)

        results_table.append({
            "name": config["save_path"].rsplit(sep='/')[7].split(".")[0],
            **res

            })
        save_results(res)
results_table
```
